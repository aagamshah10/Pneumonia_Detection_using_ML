```python
import keras
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import os
from keras.preprocessing.image import ImageDataGenerator
```

```python
mytest_route = "/content/gdrive/MyDrive/Colab Notebooks /mytest"
mytest_generator = ImageDataGenerator(rescale=1/255.0)
mytesting_set = mytest_generator.flow_from_directory(mytest_route,
                                                target_size=(224,224),batch_size=32,class_mode='categorical')
```

```python
mytest_path_pneumonia = "/content/gdrive/MyDrive/Colab Notebooks /mytest/PNEUMONIA"
mytest_path_normal = "/content/gdrive/MyDrive/Colab Notebooks /mytest/NORMAL"
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model_path = '/content/gdrive/MyDrive/Colab Notebooks/cnn_model.h5'
```

```python
img_width, img_height = 224, 224
```

```python
model = load_model(model_path)
```

```python
optm = opt = keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=optm,
              metrics=['accuracy'])
```

```python
images = []
for img in os.listdir(mytest_path_pneumonia):
    img = os.path.join(mytest_path_pneumonia , img)
    img = image.load_img(img, target_size=(img_width, img_height))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    pred = model.predict(img)
    images.append(pred)
```

```python
lst = []
for i in images:
  lst.append(i.argmax())
```

```python
imagess = []
for img in os.listdir(mytest_path_normal ):
    img = os.path.join(mytest_path_normal , img)
    img = image.load_img(img, target_size=(img_width, img_height))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    pred = model.predict(img)
    imagess.append(pred)
```

```python
lst1 = []
for i in imagess:
  lst1.append(i.argmax())
```

```python
final_list = lst1 + lst
```

```python
y_true = mytesting_set.labels
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
con_mat = confusion_matrix(final_list,y_true)
```

```python
rep = classification_report(final_list,y_true)
```
